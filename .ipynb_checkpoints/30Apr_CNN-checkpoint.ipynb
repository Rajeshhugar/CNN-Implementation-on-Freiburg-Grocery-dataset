# Data Augmentation

```python
# featurewise_center=False, 
# samplewise_center=False, 
# featurewise_std_normalization=False, 
# samplewise_std_normalization=False, 
# zca_whitening=False, 
# zca_epsilon=1e-06, 
# rotation_range=0, 
# width_shift_range=0.0, 
# height_shift_range=0.0, 
# brightness_range=None, 
# shear_range=0.0, 
# zoom_range=0.0, 
# channel_shift_range=0.0, 
# fill_mode='nearest', 
# cval=0.0, 
# horizontal_flip=False, 
# vertical_flip=False, 
# rescale=None, 
# preprocessing_function=None, 
# data_format=None, 
# validation_split=0.0, 
# dtype=None
```

```python
# Data Augmentation
train_data_generator = ImageDataGenerator(rescale = 1./255., 
                                         shear_range = 0.3, 
                                         zoom_range = 0.2, 
                                         rotation_range = 0.3, 
                                         width_shift_range = 0.2, 
                                         height_shift_range = 0.2)

train_data = train_data_generator.flow_from_directory(directory = train_path,
                                                      target_size = (200,200),
                                                      classes = classes_,
                                                      batch_size = 10,
                                                      class_mode = 'categorical')

val_data_generator = ImageDataGenerator(rescale = 1./255., 
                                         shear_range = 0.3, 
                                         zoom_range = 0.2, 
                                         rotation_range = 0.3, 
                                         width_shift_range = 0.2, 
                                         height_shift_range = 0.2)

val_data = val_data_generator.flow_from_directory(directory = val_path,
                                                  target_size = (200,200),
                                                  classes = classes_,
                                                  batch_size = 10,
                                                  class_mode = 'categorical')
```

# Modelling

```python
# 1. Model Instantiation
# 2. Feature Extractor creation
# 3. Flattening
# 4. Adding the ANN
```

```python
# Instantiating the model
cnn = tf.keras.Sequential()

#--------------------- Generating the feature extractor
# conv layer > pooling layer
# Adding the first conv layer
cnn.add(tf.keras.layers.Conv2D(input_shape = (200,200,3),
                              filters = 64, 
                              kernel_size = (3,3), 
                              padding = 'same', 
                              activation = 'relu',
                              kernel_initializer = 'he_uniform'))
# Adding the first Pooling layer
cnn.add(tf.keras.layers.MaxPool2D(pool_size = (2,2)))

# Adding the second conv layer
cnn.add(tf.keras.layers.Conv2D(filters = 128, 
                              kernel_size = (3,3), 
                              padding = 'same', 
                              activation = 'relu',
                              kernel_initializer = 'he_uniform'))
# Adding the second Pooling layer
cnn.add(tf.keras.layers.MaxPool2D(pool_size = (2,2)))

# Adding the third conv layer
cnn.add(tf.keras.layers.Conv2D(filters = 256, 
                              kernel_size = (3,3), 
                              padding = 'same', 
                              activation = 'relu',
                              kernel_initializer = 'he_uniform'))
# Adding the third Pooling layer
cnn.add(tf.keras.layers.MaxPool2D(pool_size = (2,2)))

# Adding the fourth conv layer
cnn.add(tf.keras.layers.Conv2D(filters = 256, 
                              kernel_size = (3,3), 
                              padding = 'same', 
                              activation = 'relu',
                              kernel_initializer = 'he_uniform'))
# Adding the fourth Pooling layer
cnn.add(tf.keras.layers.MaxPool2D(pool_size = (2,2)))
#--------------------- Generating the feature extractor

#--------------------- Flattening layer
cnn.add(tf.keras.layers.Flatten())
#--------------------- Flattening layer

#--------------------- Generating the Fully Connected Layers > ANN
cnn.add(tf.keras.layers.Dense(units = 1500,
                              activation = 'relu',
                              kernel_initializer = 'he_uniform'))
cnn.add(tf.keras.layers.Dense(units = 1000,
                              activation = 'relu',
                              kernel_initializer = 'he_uniform'))
cnn.add(tf.keras.layers.Dense(units = 25,
                              activation = 'softmax',
                              kernel_initializer = 'glorot_uniform'))
#--------------------- Generating the Fully Connected Layers > ANN
```

```python
cnn.compile(optimizer = 'Adam',
            loss = 'categorical_crossentropy',
            metrics = ['accuracy','Precision','Recall'])
```

```python
# Summarizing the model
cnn.summary()
```

```python
# Training the model
history_object = cnn.fit(train_data,
                         epochs = 10,
                         batch_size = 10,
                         validation_data = val_data,
                         use_multiprocessing = True,
                         workers = 30)
```
